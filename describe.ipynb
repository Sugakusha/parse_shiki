```python
import pandas as pd
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
data = pd.read_csv('proces.csv', sep = '\t', encoding = 'utf-8', index_col = 0)
data.columns
```

```python
data.groupby(['Rating', 'Shounen'])['Viewed'].sum()
```

```python
sns.lmplot(x = 'Viewed', y = 'Rated at 10', data = data)
```

```python
sns.lmplot(x = 'Viewed', y = 'Rated at 10', data = data,
           fit_reg = False, # No regression line
           hue = 'Rating')
```

```python
sns.boxplot(data = data[['Rated at 10',
       'Rated at 8', 'Rated at 9']])
```

```python
sns.violinplot(x = 'Viewed', y = 'Rated at 10', data = data)
```

```python
sns.heatmap(data[['Rated at 1', 'Rated at 10', 'Rated at 2',
       'Rated at 3', 'Rated at 4', 'Rated at 5', 'Rated at 6', 'Rated at 7',
       'Rated at 8', 'Rated at 9']].corr())
```

```python
sns.distplot(data['Viewed'])
```

```python
sns.jointplot(x = 'Viewed', y = 'Rated at 10', data = data)
```

```python
data.hist(column = ['Rated at 10',
       'Rated at 8', 'Rated at 9'],figsize = (15, 10))
```

```python
sns.pairplot(data = data['Rated at 1', 'Rated at 10', 'Rated at 2',
       'Rated at 3', 'Rated at 4', 'Rated at 5', 'Rated at 6', 'Rated at 7',
       'Rated at 8', 'Rated at 9'], hue = 'Rating')
```
